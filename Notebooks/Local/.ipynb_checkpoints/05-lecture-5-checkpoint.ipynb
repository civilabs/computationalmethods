## Procedure for using FDM in 2D flow in soils

***Step 1:***
- Divide the flow domain into a square grid.
- Generally, finer grids give more accurate solution than coarser grids.
- If the flow is symmetrical, you only need to consider one-half of the flow domain (below is an example).

***Step 2:***
- Identify boundary conditions, example, impermeable boundaries (flow lines) and permeable boundaries (equipotential lines).

<br/><div></div>

***Step 3:***
- Determine the heads at the permeable or equipotential boundaries. For example, the heads along the equipotential boundary AB (figure above) is Î”H.
- Therefore, all the nodes along this boundary will have a constant head of Î”H.
- Because of symmetry, the head along nodes directly under the sheet pile wall (EF) is Î”H/2.

***Step 4:***
- Apply the known heads to the corresponding nodes and assume reasonable initial values for the interior nodes.

***Step 5:***
- Apply Eqn. (5.1) if the soil is isotropic to each node except: 
        a. at impermeable boundaries â€“ eqn. (5.2)
        b. at corners â€“ eqn. (5.3) & (5.4)
        c. at nodes where the heads are known

***Step 6:***
- Repeat item 5 until the new value at the node differs from the old value by a small numerical tolerance for example 0.001 m.

***Step 7:***
- Arbitrarily select a sequential set of nodes along a column of nodes and calculate the flow, q, using eqn. (5.6). It is best to calculate qâ€™ = q for a unit permeability value to avoid too many decimal points in the calculations.

***Step 8:***
- Repeat items 1 to 6, to find the flow distribution by replacing heads by flow qâ€™. For example, the flow rate, calculated in item 7 is applied to all nodes along AC and CF.

***Step 9:***
- Calculate the pore water pressure distribution using eqn. (5.5).

### Step 4
Insert the heads at the nodes.

```python
# Initialize head with ones
h = np.ones((7,13))

# Assign head of 3 to nodes on AB
h[0,:] = 3

# Assign head of 1.5 to nodes on CD
h[3:7,12] = 1.5
```

### Step 5 and Step 6
Apply the appropriate equations and solve.

```python
# Apply the appropriate equations and solve
for itr in range(1,1000):
    # Boundary AE
    for j in range(1,6):
        h[j,0] = (h[j-1,0] + h[j+1,0] + 2*h[j,1]) / 4
    h[6,0] = (h[5,0] + h[6,1]) / 2
    
    # Boundary ED
    for i in range(1,12):
        h[6,i] = (h[6,i-1] + h[6,i+1] + 2*h[5,i]) / 4
        
    # Boundary BC
    for j in range(1,3):
        h[j,12] = (h[j-1,12] + h[j+1,12] + 2*h[j,11]) / 4
    
    # Remaining intermediate nodes
    for j in range(1,6):
        for i in range(1,12):
            h[j,i] = (h[j-1,i] + h[j+1,i] + h[j,i+1] + h[j,i-1]) / 4
```

### Formatting data (optional)

```python
# Defining row and column names
row = ['0 m', '2 m', '4 m', '6 m', '8 m', '10 m', '12 m']
col = ['0 m', '2 m', '4 m', '6 m', '8 m', '10 m', '12 m', '14 m', '16 m', '18 m', '20 m', '22 m', '24 m']

# Create dataframe for head data
df_h = pd.DataFrame(np.around(h,3), row, col)

# View data frame
df_h
```

### Plotting Head Contours (Equipotential Lines)

```python
# Preparing mesh grid and plotting area
x_list = np.linspace(0, 24, 13)
y_list = np.linspace(0, 12, 7)
X, Y = np.meshgrid(x_list, y_list)
fig,ax=plt.subplots(1,1)

# Plotting contours (equipotential lines)
cp = ax.contour(X, Y, h)
# fig.colorbar(cp)

# Setting title
ax.set_title('Equipotential Lines')

# Rearranging axes as per our requirement
ax.invert_yaxis()
ax.xaxis.set_ticks_position('top')
```

### Step 7
Calculate q.

```python
# Apply the appropriate equations and solve
for itr in range(1,1000):
    # Boundary AE
    for j in range(1,6):
        h[j,0] = (h[j-1,0] + h[j+1,0] + 2*h[j,1]) / 4
    h[6,0] = (h[5,0] + h[6,1]) / 2
    
    # Boundary ED
    for i in range(1,12):
        h[6,i] = (h[6,i-1] + h[6,i+1] + 2*h[5,i]) / 4
        
    # Boundary BC
    for j in range(1,3):
        h[j,12] = (h[j-1,12] + h[j+1,12] + 2*h[j,11]) / 4
    
    # Remaining intermediate nodes
    for j in range(1,6):
        for i in range(1,12):
            h[j,i] = (h[j-1,i] + h[j+1,i] + h[j,i+1] + h[j,i-1]) / 4
```
