# 1. Introduction

In FDM derivatives in the governing equations are written in finite difference equations. To illustrate, let us consider the second-order, one-dimensional differential equation:

$$ {d^{2}u\over dx^{2}}-2=0, 0<x<1 $$

With Dirichlet boundary conditions (values of variable u specified at boundaries): u=0 at x=0 and u=0 at x=1, the exact Solution is:

$$ u=x^2 - x $$

In FDM, solutions are computed at specific - discrete – points called Nodes. The process of subdividing the domain into discrete points is called discretization. Starting from x=0 the domain is subdivided in steps, until x=1. The chosen step is called step size and is denoted by Δx.

<br/> <div><img src="https://raw.githubusercontent.com/civilabs/computationalmethods/main/Others/Images/Lecture2/4.png" width="500"/></div>

Finite Difference Equations:

Forward difference

$$ ({du \over dx})_i ≈ {u_{i+1} - u_{i} \over Δx} $$

Backward difference

$$ ({du \over dx})_i ≈ {u_{i} - u_{i-1} \over Δx} $$

Central difference

$$ ({du \over dx})_i ≈ {u_{i+1} - u_{i-1} \over 2Δx} $$

## Code Solution

### Importing required libraries

```python
# Import required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['figure.dpi'] = 144
```

### Computing porewater pressure

```python
# Initialize pore water pressure with zeros
u = np.zeros((6,6))

# Assign initial pore water pressure values
u[:,0] = [0,78,72,62,48,30]

# Calculate alpha with Cv = 2.5 m2/year
a = 2.5 * 0.1 / 1**2

# Calculate pore water pressure values
for j in range(0,5):
    for i in range(1,5):
        u[i,j+1] = u[i,j] + a * (u[i+1,j] - 2 * u[i,j] + u[i-1,j])
        u[5,j+1] = u[5,j] + a * (2 * u[5-1,j] - 2 * u[5,j])
        
# View the final result
np.around(u,3)
```

### Formatting data (optional)

```python
# Defining row and column names
row = ['0 m', '1 m', '2 m', '3 m', '4 m', '5 m']
col = ['0 yrs', '0.1 yrs', '0.2 yrs', '0.3 yrs', '0.4 yrs', '0.5 yrs']

# Create dataframe for pore water pressure data
df_u = pd.DataFrame(np.around(u,3), row, col)

# View data frame
df_u
```

### Plotting result

```python
# Add plot
plt.plot(u,row,label=col)

# Add legend
plt.legend(loc="upper right")

# Make x axis from 0 to 110
plt.xlim(0,110)

# Get current axis
ax = plt.gca()

# Invert y axis so that the depth is in reverse order
ax.invert_yaxis()

# Make x axis label and ticks at the top
ax.xaxis.set_ticks_position('top')
ax.xaxis.set_label_position('top')

# Set x and y labels
ax.set_xlabel('Excess pore water pressure')
ax.set_ylabel('Depth')
```
