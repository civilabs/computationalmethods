# 1. Consolidation Settlement and Average Degree of Consolidation - FD Solution

Conventional consolidation settlement analysis using empirical equations has the following limitations:
1. The coefficient of consolidation Cv  is assumed to be constant, however, it is well known that Cv changes with time in the consolidation process.
2. Variation in application of load such as stage loading (load increasing by a certain rate), and abrupt change of load (the case of surcharge preloading) are not considered.

Numerical solution using the finite difference method can be used to overcome the problem. The numerical solution does not require the empirical correlation for Tv because it can be done by using the original formula and load application and other variations can be modeled almost correctly.

In Eqn. (A), the integral of the excess pore pressure cannot be evaluated exactly because the excess pore pressure is only at the grid points. However, the integral can be evaluated approximately using some numerical techniques. The simplest approach is to use the Trapezoidal Method.

## Variable Loading

In reality, a loading is never placed instantaneously and the problem can be easily handled through finite difference formulation. Variable loading usually occurs when there is a structure being built on top of the soil layer. As the construction of the structure proceeds, the load induced onto the ground gradually increases with time.Considering that the construction is a time consuming process, we cannot assume a formulation of instantaneous load to be used in the calculation of settlement.

## Abrupt Change of Load

When the existing soil layers are too weak, it has to be loaded in stages. Loading in stages will increase the shear strength and bearing capacity of the soil as it consolidates.

## Variable Coefficient of Consolidation (Cv)

The coefficient of consolidation is usually considered as constant in consolidation analysis although it is a variable quantity. It is known empirically that CV changes during consolidation as the void ratio of the soil changes causing a decrease in permeability and compressibility of the soil.

## Code Solution

### Importing required libraries

```python
# Import required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['figure.dpi'] = 144
```

### *1. Case 1: Constant load scenario: a uniform load of 90 kPa is applied onto the soil layer at once*

#### Computing porewater pressure

```python
# Initialize pore water pressure with zeros
u = np.zeros((8,11))

# Assign initial pore water pressure values
u[:,0] = [0,0,0,90,90,90,90,0]
u[1,:] = 90

# Calculate alpha with Cv = 2 m2/year
a = 2 * 0.2 / 1**2

# Calculate pore water pressure values
for j in range(0,10):
    for i in range(3,6):
        u[i,j+1] = u[i,j] + a * (u[i+1,j] - 2 * u[i,j] + u[i-1,j])
        u[6,j+1] = u[6,j] + a * (2 * u[6-1,j] - 2 * u[6,j])
        
# Calculate degree of consolidation
int_t_0 = ((90 + 90)/2 + sum([90,90,90])) * 1
for j in range(1,11):
    int_t_j = ((u[2,j] + u[6,j])/2 + sum([u[3,j],u[4,j],u[5,j]])) * 1
    u[7,j] = (1 - int_t_j/int_t_0) * 100
    
# Calculate settlement for mv of 0.00083 m2/kN
mv = 0.00083
for j in range(1,11):
    int_t_j = ((u[2,j] + u[6,j])/2 + sum([u[3,j],u[4,j],u[5,j]])) * 1
    u[0,j] = mv * (90 * 4 - int_t_j) * 1000
        
# View the final result
np.around(u,1)
```

#### Formatting data (optional)

```python
# Defining row and column names
row = ['Settlement (mm)', 'q (kPa)', '0 m', '1 m', '2 m', '3 m', '4 m', 'Uav (%)']
col = ['0 yrs', '0.2 yrs', '0.4 yrs', '0.6 yrs', '0.8 yrs', '1 yrs', '1.2 yrs', '1.4 yrs', '1.6 yrs', '1.8 yrs', '2 yrs']

# Create dataframe for pore water pressure data
df_u = pd.DataFrame(np.around(u,3), row, col)

# View data frame
df_u
```

#### Plotting result

```python
# Add plot
plt.plot(u[2:7,:],row[2:7], label=col)

# Add legend
plt.legend(loc="upper right")

# Make x axis from 0 to 110
plt.xlim(0,120)

# Get current axis
ax = plt.gca()

# Invert y axis so that the depth is in reverse order
ax.invert_yaxis()

# Make x axis label and ticks at the top
ax.xaxis.set_ticks_position('top')
ax.xaxis.set_label_position('top')

# Set x and y labels
ax.set_xlabel('Excess pore water pressure')
ax.set_ylabel('Depth');
```

### *2. Case 2: Variable load scenario: a fill having a unit weight 15 kN/m3 is placed at a rate of 0.5 m/month during the first year and the filling stops after one year. The final load would be 90 kPa.*

#### Computing porewater pressure

```python
# Initialize pore water pressure with zeros
u = np.zeros((8,21))

# Assign initial pore water pressure values
u[1,1:10] = [9, 18, 27, 36, 45, 54, 63, 72, 81]
u[1,10:21] = 90

# Calculate alpha with Cv = 2 m2/year
a = 2 * 0.1 / 1**2

# Calculate pore water pressure values until the load increase ends
for j in range(0,10):
    for i in range(3,6):
        u[i,j+1] = u[i,j] + a * (u[i+1,j] - 2 * u[i,j] + u[i-1,j]) + 9
        u[6,j+1] = u[6,j] + a * (2 * u[6-1,j] - 2 * u[6,j]) + 9
        
# Calculate pore water pressure values after the load increase ends
for j in range(10,20):
    for i in range(3,6):
        u[i,j+1] = u[i,j] + a * (u[i+1,j] - 2 * u[i,j] + u[i-1,j])
        u[6,j+1] = u[6,j] + a * (2 * u[6-1,j] - 2 * u[6,j])
        
# Calculate degree of consolidation
for j in range(1,21):
    int_t_0 = ((u[1,j] + u[1,j])/2 + sum([u[1,j],u[1,j],u[1,j]])) * 1
    int_t_j = ((u[2,j] + u[6,j])/2 + sum([u[3,j],u[4,j],u[5,j]])) * 1
    u[7,j] = (1 - int_t_j/int_t_0) * 100
    
# Calculate settlement for mv of 0.00083 m2/kN
mv = 0.00083
for j in range(1,21):
    int_t_j = ((u[2,j] + u[6,j])/2 + sum([u[3,j],u[4,j],u[5,j]])) * 1
    u[0,j] = mv * (u[1,j] * 4 - int_t_j) * 1000
        
# View the final result
np.around(u)
```

#### Formatting data (optional)

```python
# Defining row and column names
row = ['Settlement (mm)', 'q (kPa)', '0 m', '1 m', '2 m', '3 m', '4 m', 'Uav (%)']
col = [str(round(i,1)) + ' yrs' for i in np.arange(0,2.1,0.1)]

# Create dataframe for pore water pressure data
df_u = pd.DataFrame(np.around(u,3), row, col)

# View data frame
df_u
```

#### Plotting result

```python
# Add plot
plt.plot(u[2:7,:],row[2:7], label=col)

# Add legend
plt.legend(loc="upper right", prop={'size': 6})

# Make x axis from 0 to 110
plt.xlim(0,120)

# Get current axis
ax = plt.gca()

# Invert y axis so that the depth is in reverse order
ax.invert_yaxis()

# Make x axis label and ticks at the top
ax.xaxis.set_ticks_position('top')
ax.xaxis.set_label_position('top')

# Set x and y labels
ax.set_xlabel('Excess pore water pressure')
ax.set_ylabel('Depth');
```

### *3. Case 3: Abrupt change of load scenario: In the abrupt change of load scenario, the soil layer will be loaded with fill of 60 kPa and after one year, an additional fill of 30 kPa will be laid on the existing fill.*

#### Computing porewater pressure

```python
# Initialize pore water pressure with zeros
u = np.zeros((8,11))

# Assign initial pore water pressure values
u[:,0] = [0,0,0,60,60,60,60,0]
u[1,0:6] = 60
u[1,6:11] = 90

# Calculate alpha with Cv = 2 m2/year
a = 2 * 0.2 / 1**2

# Calculate pore water pressure values until the abrupt change
for j in range(0,5):
    for i in range(3,6):
        u[i,j+1] = u[i,j] + a * (u[i+1,j] - 2 * u[i,j] + u[i-1,j])
        u[6,j+1] = u[6,j] + a * (2 * u[6-1,j] - 2 * u[6,j])
        
# Calculate pore water pressure values at the abrupt change
for i in range(3,6):
    u[i,6] = u[i,5] + a * (u[i+1,5] - 2 * u[i,5] + u[i-1,5]) + 30
    u[6,6] = u[6,5] + a * (2 * u[6-1,5] - 2 * u[6,5]) + 30
        
# Calculate pore water pressure values after the abrupt change
for j in range(6,10):
    for i in range(3,6):
        u[i,j+1] = u[i,j] + a * (u[i+1,j] - 2 * u[i,j] + u[i-1,j])
        u[6,j+1] = u[6,j] + a * (2 * u[6-1,j] - 2 * u[6,j])
        
# Calculate degree of consolidation
for j in range(1,11):
    int_t_0 = ((u[1,j] + u[1,j])/2 + sum([u[1,j],u[1,j],u[1,j]])) * 1
    int_t_j = ((u[2,j] + u[6,j])/2 + sum([u[3,j],u[4,j],u[5,j]])) * 1
    u[7,j] = (1 - int_t_j/int_t_0) * 100
    
# Calculate settlement for mv of 0.00083 m2/kN
mv = 0.00083
for j in range(1,11):
    int_t_j = ((u[2,j] + u[6,j])/2 + sum([u[3,j],u[4,j],u[5,j]])) * 1
    u[0,j] = mv * (u[1,j] * 4 - int_t_j) * 1000
        
# View the final result
np.around(u)
```

#### Formatting data (optional)

```python
# Defining row and column names
row = ['Settlement (mm)', 'q (kPa)', '0 m', '1 m', '2 m', '3 m', '4 m', 'Uav (%)']
col = [str(round(i,1)) + ' yrs' for i in np.arange(0,2.1,0.2)]

# Create dataframe for pore water pressure data
df_u = pd.DataFrame(np.around(u,3), row, col)

# View data frame
df_u
```

#### Plotting result

```python
# Add plot
plt.plot(u[2:7,:],row[2:7], label=col)

# Add legend
plt.legend(loc="upper right")

# Make x axis from 0 to 110
plt.xlim(0,120)

# Get current axis
ax = plt.gca()

# Invert y axis so that the depth is in reverse order
ax.invert_yaxis()

# Make x axis label and ticks at the top
ax.xaxis.set_ticks_position('top')
ax.xaxis.set_label_position('top')

# Set x and y labels
ax.set_xlabel('Excess pore water pressure')
ax.set_ylabel('Depth');
```

### *4. Case 4: Constant Load with Variable Cv: the Cv is assumed to decrease with a rate of 0.1 m2/year.constant Load with Variable Cv: the Cv is assumed to decrease with a rate of 0.1 m2/year.*

#### Computing porewater pressure

```python
# Initialize pore water pressure with zeros
u = np.zeros((9,11))

# Assign initial pore water pressure values
u[:,0] = [0,0,0,0,90,90,90,90,0]
u[1,:] = 90

# Assign Cv values
u[2,:] = [i for i in np.arange(2,1.79,-0.02)]

# Calculate pore water pressure values
for j in range(0,10):
    # Calculate alpha
    a = u[2,j+1] * 0.2 / 1**2
    for i in range(4,7):
        u[i,j+1] = u[i,j] + a * (u[i+1,j] - 2 * u[i,j] + u[i-1,j])
        u[7,j+1] = u[7,j] + a * (2 * u[7-1,j] - 2 * u[7,j])
        
# Calculate degree of consolidation
int_t_0 = ((90 + 90)/2 + sum([90,90,90])) * 1
for j in range(1,11):
    int_t_j = ((u[3,j] + u[7,j])/2 + sum([u[4,j],u[5,j],u[6,j]])) * 1
    u[8,j] = (1 - int_t_j/int_t_0) * 100
    
# Calculate settlement for mv of 0.00083 m2/kN
mv = 0.00083
for j in range(1,11):
    int_t_j = ((u[3,j] + u[7,j])/2 + sum([u[4,j],u[5,j],u[6,j]])) * 1
    u[0,j] = mv * (90 * 4 - int_t_j) * 1000
        
# View the final result
np.around(u)
```

#### Formatting data (optional)

```python
# Defining row and column names
row = ['Settlement (mm)', 'q (kPa)', 'Cv (m2/yr)', '0 m', '1 m', '2 m', '3 m', '4 m', 'Uav (%)']
col = [str(round(i,1)) + ' yrs' for i in np.arange(0,2.1,0.2)]

# Create dataframe for pore water pressure data
df_u = pd.DataFrame(np.around(u,3), row, col)

# View data frame
df_u
```

#### Plotting result

```python
# Add plot
plt.plot(u[3:8,:],row[3:8], label=col)

# Add legend
plt.legend(loc="upper right")

# Make x axis from 0 to 110
plt.xlim(0,120)

# Get current axis
ax = plt.gca()

# Invert y axis so that the depth is in reverse order
ax.invert_yaxis()

# Make x axis label and ticks at the top
ax.xaxis.set_ticks_position('top')
ax.xaxis.set_label_position('top')

# Set x and y labels
ax.set_xlabel('Excess pore water pressure')
ax.set_ylabel('Depth');
```
